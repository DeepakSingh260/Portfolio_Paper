```python
import pandas as pd 
```

```python
data = pd.read_csv("final_enriched_data.csv")
```

```python
data.dropna(subset=['Quarterly_Ticker'])
```

```python
data.sort_values(by=["Quarterly_Ticker" , "Date"],inplace=True)
```

```python
data.shape
```

```python
data["Price_Change"] = data.groupby("Quarterly_Ticker")["Close"].pct_change()

data["Rolling_Mean"] = data.groupby("Quarterly_Ticker")["Close"].transform(lambda x: x.rolling(window=5).mean())
#target Variable
data["Future_Price_Change"] = data.groupby("Quarterly_Ticker")["Close"].shift(-1) / data["Close"] - 1


data["Oil_Price_Change"] = data.groupby("Quarterly_Ticker")["Oil Price"].pct_change()
data["Gas_Price_Change"] = data.groupby("Quarterly_Ticker")["Gas Price"].pct_change()
data["Gold_Price_Change"] = data.groupby("Quarterly_Ticker")["Gold Price"].pct_change()
```

```python
data.to_csv("modified_data")
```
